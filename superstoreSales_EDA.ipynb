# Data Cleaning

```python
df.columns
```

```python
#Splitting the date and time
df[["Shipping.Date", "Shipping.Time"]] = df["Ship.Date"].str.split(' ', 1, expand=True)
df[["Order_Date", "Order_Time"]] = df["Order.Date"].str.split(' ', 1, expand=True)

#splitting the Order_Date into year, month, date for future analysis
df[['year','month','date']] = df["Order_Date"].str.split('-', 2, expand=True)
df.head()
```

```python
#checking to see if the columns have unique values to see which to drop
df["Market2"].unique()
df["Discount"].unique()
df["记录数"].unique() # does not have unique values(contains only 1)
```

```python
# Dropping columns we don't need
df = df.drop(columns = ["Order.Date","Ship.Date","Shipping.Time", "Order_Time","记录数","date"])
df.head()
```

# Exploratory Data Analysis

### Sales Performance Insights

#### Total Sales and Profit by Category and Sub.Category

```python
df.groupby(['Category','Sub.Category',]).agg({'Sales': ['sum']})
```

```python
# grouping the top sales by category
cat_sales = df.groupby('Category').sum()
cat_sales = cat_sales.filter(items = ['Sales','Profit'])
cat_sales.sort_values(by='Sales', ascending=False)
```

```python
# Plotting Sales and Profit for each Category
fig, ax1 = plt.subplots(figsize=(5, 3))

# Plot Sales
ax1.bar(cat_sales.index, cat_sales['Sales'], color='purple', label='Sales')
ax1.set_xlabel('Category')
ax1.set_ylabel('Sales', color='purple')
ax1.tick_params(axis='y', labelcolor='purple')

# Create another axis for Profit
ax2 = ax1.twinx()
ax2.plot(cat_sales.index, cat_sales['Profit'], color='blue', marker='o', label='Profit')
ax2.set_ylabel('Profit', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# Title and layout
plt.title('Sales and Profit by Category')
fig.tight_layout()
plt.xticks(rotation=90)
plt.show()
```

```python
# grouping the top sales by sub_category
subCat_sales = df.groupby('Sub.Category').sum()
subCat_sales = subCat_sales.filter(items = ['Sales','Profit'])
subCat_sales.sort_values(by='Sales', ascending=False)
```

```python
# Plotting Sales and Profit for each Sub.Category
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot Sales
ax1.bar(subCat_sales.index, subCat_sales['Sales'], color='purple', label='Sales')
ax1.set_xlabel('Sub.Category')
ax1.set_ylabel('Sales', color='purple')
ax1.tick_params(axis='y', labelcolor='purple')
ax1.set_xticklabels(subCat_sales.index, rotation=45, ha='right')

# Create another axis for Profit
ax2 = ax1.twinx()
ax2.plot(subCat_sales.index, subCat_sales['Profit'], color='lightgreen', marker='o', label='Profit')
ax2.set_ylabel('Profit', color='lightgreen')
ax2.tick_params(axis='y', labelcolor='lightgreen')

# Title and layout
plt.title('Sales and Profit by Sub.Category')
fig.tight_layout()
plt.xticks(rotation=90)
plt.show()
```

### Customer Insights

#### Top Customers by Sales and Profit

```python
# top 5 customers by sales
sales_by_customer = df.groupby('Customer.Name')['Sales'].sum()
sales_by_customer = sales_by_customer.sort_values( ascending=False)
sales_by_customer = sales_by_customer[:5]
sales_by_customer
```

```python
# Define a function to display sales numbers on the pie chart
def func(pct, allvalues):
    absolute = int(pct / 100. * sum(allvalues))
    return f'{absolute}'

# Plotting the pie chart
plt.figure(figsize=(10, 6))
plt.pie(sales_by_customer, labels=sales_by_customer.index, autopct=lambda pct: func(pct, sales_by_customer), startangle=140)
plt.title('Total Customers by Sales')
plt.axis('equal') 
plt.show()
```

```python
# top 5 customers by profit
profit_by_customer = df.groupby('Customer.Name')['Profit'].sum()
profit_by_customer = profit_by_customer.sort_values( ascending=False)
profit_by_customer = profit_by_customer[:5]
profit_by_customer
```

```python
# Define a function to display Profit numbers on the pie chart
def func(pct, allvalues):
    absolute = int(pct / 100. * sum(allvalues))
    return f'{absolute}'

# Plotting the pie chart
plt.figure(figsize=(10, 6))
plt.pie(profit_by_customer, labels=profit_by_customer.index, autopct=lambda pct: func(pct, profit_by_customer), startangle=140)
plt.title('Total Customers by Profit')
plt.axis('equal') 
plt.show()
```

### Market Analysis

#### Market and Market2 Performance

```python
#Market and Market2 Performance
Market = df.groupby('Market')['Sales','Profit','Quantity'].sum()
Market2 = df.groupby('Market2')['Sales','Profit','Quantity'].sum()
print(Market, Market2)
```

```python
# Ensure the x-axis matches the length of the data
x1 = np.arange(len(Market.index))  # Create an array for the x-axis positions for Market
x2 = np.arange(len(Market2.index))  # Create an array for Market2

# Set the width of the bars
width = 0.2

# Create the plot with two subplots side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))  # 1 row, 2 columns

# --- First subplot (Market) ---
# Plot each bar for 'Sales', 'Profit', and 'Quantity', slightly shifting them to create a grouped effect
ax1.bar(x1 - width, Market['Sales'], width=width, color='green', label='Sales')
ax1.bar(x1, Market['Profit'], width=width, color='blue', label='Profit')
ax1.bar(x1 + width, Market['Quantity'], width=width, color='orange', label='Quantity')

# Set x-ticks to be the year labels from the index of the DataFrame
ax1.set_xticks(x1)
ax1.set_xticklabels(Market.index)

# Add labels and title
ax1.set_xlabel('Market')
ax1.set_ylabel('Values')
ax1.set_title('Sales, Profit, and Quantity of Market')

# Add legend
ax1.legend()

# --- Second subplot (Market2) ---
# Plot each bar for 'Sales', 'Profit', and 'Quantity', slightly shifting them to create a grouped effect
ax2.bar(x2 - width, Market2['Sales'], width=width, color='green', label='Sales')
ax2.bar(x2, Market2['Profit'], width=width, color='blue', label='Profit')
ax2.bar(x2 + width, Market2['Quantity'], width=width, color='orange', label='Quantity')

# Set x-ticks to be the year labels from the index of the DataFrame
ax2.set_xticks(x2)
ax2.set_xticklabels(Market2.index)

# Add labels and title
ax2.set_xlabel('Market2')
ax2.set_ylabel('Values')
ax2.set_title('Sales, Profit, and Quantity of Market2')

# Add legend
ax2.legend()

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

### Discounts and Profitability

#### Effect of Discounts on Profit over the year

```python
sales_sum = df.groupby('year')['Profit'].sum().reset_index()
discount_avg = df.groupby('year')['Discount'].mean().reset_index()
discount_on_profit_year = sales_sum.merge(discount_avg)
```

```python
# Create the plot
fig, ax1 = plt.subplots(figsize=(10, 5))

# Plot Profit
ax1.scatter(discount_on_profit_year['year'], discount_on_profit_year['Profit'], color='green', label='Profit')
ax1.set_xlabel('Year')
ax1.set_ylabel('Profit', color='green')
ax1.tick_params(axis='y', labelcolor='green')

# Create a second y-axis for Discount
ax2 = ax1.twinx()
ax2.scatter(discount_on_profit_year['year'], discount_on_profit_year['Discount'], color='orange', label='Discount')
ax2.set_ylabel('Discount', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

# Add title and legend
plt.title('Correlation between Discount and Profit over year')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Show the plot
plt.show()
```

#### Effect of Discounts on Profit over the month

```python
sales_sum = df.groupby('month')['Profit'].sum().reset_index()
discount_avg = df.groupby('month')['Discount'].mean().reset_index()
discount_on_profit_month = sales_sum.merge(discount_avg)
```

```python
# Create the plot
fig, ax1 = plt.subplots(figsize=(10, 5))

# Plot Profit
ax1.scatter(discount_on_profit_month['month'], discount_on_profit_month['Profit'], color='green', label='Profit')
ax1.set_xlabel('Month')
ax1.set_ylabel('Profit', color='green')
ax1.tick_params(axis='y', labelcolor='green')

# Create a second y-axis for Discount
ax2 = ax1.twinx()
ax2.scatter(discount_on_profit_month['month'], discount_on_profit_month['Discount'], color='orange', label='Discount')
ax2.set_ylabel('Discount', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

# Add title and legend
plt.title('Correlation between Discount and Profit over month')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Show the plot
plt.show()
```
